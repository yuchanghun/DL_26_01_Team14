# 01. 샘플 데이터 생성
실제 K-Fashion 데이터 도착 전 파이프라인 테스트용 CSV를 생성합니다.  
- **상품목록.csv** : 500개 상품 (23 스타일 / 10 대분류 / 색상·소재·핏 등)
- **판매데이터.csv** : 상품별 7일·30일 판매량 및 조회수

```python
import pandas as pd
import numpy as np
import os

np.random.seed(42)
os.makedirs('data', exist_ok=True)
```

```python
# K-Fashion 실제 클래스 정의 (학습된 모델과 동일)
STYLES = [
    '기타', '레트로', '로맨틱', '리조트', '매니시',
    '모던', '밀리터리', '섹시', '소피스트케이티드', '스트리트',
    '스포티', '아방가르드', '오리엔탈', '웨스턴', '젠더리스',
    '컨트리', '클래식', '키치', '톰보이', '펑크',
    '페미닌', '프레피', '히피', '힙합',
]
CATEGORIES = ['상의', '하의', '아우터', '원피스', '점프수트']
SIZES     = ['XS', 'S', 'M', 'L', 'XL']
COLORS    = ['블랙', '화이트', '네이비', '베이지', '그레이', '레드', '카키', '핑크', '블루', '브라운']
MATERIALS = ['면', '폴리에스터', '데님', '울', '린넨', '가죽', '시폰', '니트', '벨벳', '실크']
FITS      = ['레귤러핏', '오버핏', '슬림핏', '루즈핏']
LENGTHS   = ['미니', '미디', '맥시', '크롭']

print(f'스타일 {len(STYLES)}개 / 대분류 {len(CATEGORIES)}개')
```

```python
# 상품 목록 생성
products = []
for pid in range(1, 501):
    products.append({
        'product_id':   pid,
        'product_name': f'상품_{pid:03d}',
        'style':        np.random.choice(STYLES),
        'category':     np.random.choice(CATEGORIES),
        'size':         np.random.choice(SIZES),
        'color':        np.random.choice(COLORS),
        'material':     np.random.choice(MATERIALS),
        'fit':          np.random.choice(FITS),
        'length':       np.random.choice(LENGTHS),
        'price':        int(np.random.choice([19900, 29900, 39900, 49900, 69900, 89900, 119900])),
    })

df_products = pd.DataFrame(products)
df_products.to_csv('data/상품목록.csv', index=False, encoding='utf-8-sig')
print(f'상품목록.csv 저장 완료 ({len(df_products)}개)')
df_products.head()
```

```python
# 판매 데이터 생성
popular_ids = np.random.choice(range(1, 501), size=60, replace=False)
sales = []
for pid in range(1, 501):
    base = 200 if pid in popular_ids else 20
    sales.append({
        'product_id': pid,
        'sales_30d':  max(0, int(np.random.normal(base,       base * 0.3))),
        'sales_7d':   max(0, int(np.random.normal(base * 0.3, base * 0.1))),
        'view_count': max(0, int(np.random.normal(base * 5,   base))),
    })

df_sales = pd.DataFrame(sales)
df_sales.to_csv('data/판매데이터.csv', index=False, encoding='utf-8-sig')
print(f'판매데이터.csv 저장 완료 ({len(df_sales)}개)')
df_sales.describe()
```
